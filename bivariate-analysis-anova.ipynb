```python
import seaborn as sns
import matplotlib.pyplot as plt

def create_violin_plot(data, x_col, y_col, **args):
    """
    Create a violin plot to describe the distribution of x for a given categorical y.

    Parameters:
    - data: DataFrame, the dataset containing the variables.
    - x_col: str, the column name for the numerical variable (x).
    - y_col: str, the column name for the categorical variable (y).
    - title: str, the title of the plot.
    - x_label: str, the label for the x-axis.
    - y_label: str, the label for the y-axis.
    """
    fig, ax = plt.subplots(ncols=1, figsize=(100, 24), sharey=True)
    sns.violinplot(x=data[x_col], y=data[y_col],ax=ax, log_scale=True,**args)
    plt.xlabel(x_col,size=20)
    plt.ylabel(y_col,size=20)
    plt.yscale("log")

    plt.show()

# Example usage:
# Assuming you have a DataFrame named 'df' with columns 'numerical_variable' and 'categorical_variable'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def create_box_chart(data, x_col,y_col, **args):
    """
    Create a bar chart to visualize the distribution of a categorical variable.

    Parameters:
    - data: DataFrame, the dataset containing the variable.
    - x_col: str, the column name for the categorical variable.
    - title: str, the title of the plot.
    - x_label: str, the label for the x-axis.
    - y_label: str, the label for the y-axis.
    """
    plt.figure(figsize=(25, 12))  # Adjust the figure size as needed
    sns.boxplot(x=data[x_col],y=data[y_col],**args,whis=1.5)
    plt.xlabel(x_col)
    plt.yscale("log")
    plt.ylim=[100,10000]
    plt.ylabel(y_col)
    plt.show()

# Example usage:
# Assuming you have a DataFrame named 'df' with a column 'categorical_variable'
```

```python
from scipy.stats import f_oneway

def one_way_anova(*args, alpha=0.05):
    """
    Perform a one-way ANOVA test on multiple samples.

    Parameters:
    - args: arrays or iterable of arrays, each representing a sample.
    - alpha: float, significance level for the test (default is 0.05).

    Returns:
    - Tuple containing the F-statistic and p-value.
    """
    # Perform one-way ANOVA
    statistic, p_value = f_oneway(*args)

    # Print the results
    print("One-way ANOVA Results:")
    print(f"  F-statistic: {statistic}")
    print(f"  p-value: {p_value}")

    # Check for significance
    if p_value < alpha:
        print("The difference between the means is statistically significant (reject the null hypothesis).")
    else:
        print("The difference between the means is not statistically significant (fail to reject the null hypothesis).")

    return statistic, p_value

# Example usage:
# Assuming you have three samples named sample1, sample2, and sample3
sample1 = [2, 4, 6, 8, 10]
sample2 = [1, 3, 5, 7, 9]
sample3 = [3, 6, 9, 12, 15]

```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from itertools import combinations

def one_way_anova_heatmap(*args, alpha=0.05):
    """
    Perform a one-way ANOVA on all pairs of samples and present the results in a seaborn heatmap.

    Parameters:
    - args: arrays or iterable of arrays, each representing a sample.
    - alpha: float, significance level for the test (default is 0.05).

    Returns:
    - None (displays the heatmap).
    """
    num_samples = len(args)
    p_values_matrix = np.ones((num_samples, num_samples))  # Matrix for p-values

    # Perform one-way ANOVA for all pairs of samples
    for i, j in combinations(range(num_samples), 2):
        sample1, sample2 = args[i], args[j]
        _, p_value = f_oneway(sample1, sample2)

        # Store p-value in the matrix
        p_values_matrix[i, j] = p_value
        p_values_matrix[j, i] = p_value  # Since the matrix is symmetric

    # Create a DataFrame for the heatmap
    columns = [f"Sample {i+1}" for i in range(num_samples)]
    p_values_df = pd.DataFrame(p_values_matrix, index=columns, columns=columns)

    # Plot the heatmap
    plt.figure(figsize=(50, 24))  # Adjust the figure size as needed
    sns.heatmap(p_values_df, annot=True, cmap="coolwarm", vmin=0, vmax=alpha, fmt=".3f")
    plt.title("One-way ANOVA P-Values Heatmap")
    plt.show()

# Example usage:
# Assuming you have three samples named sample1, sample2, and sample3
sample1 = [2, 4, 6, 8, 10]
sample2 = [1, 3, 5, 7, 9]
sample3 = [3, 6, 9, 12, 15]

one_way_anova_heatmap(sample1, sample2, sample3)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from itertools import combinations

def one_way_anova_heatmap_from_dataframe(data, category_col, value_column,alpha=0.05,**args):
    """
    Perform a one-way ANOVA on all pairs of samples split by a categorical variable and
    present the results in a seaborn heatmap.

    Parameters:
    - data: DataFrame, the dataset containing the variables.
    - category_col: str, the column name for the categorical variable.
    - alpha: float, significance level for the test (default is 0.05).

    Returns:
    - None (displays the heatmap).
    """
    # Get unique categories in the specified column
    categories = data[category_col].unique()

    # Create an empty matrix for p-values
    num_categories = len(categories)
    p_values_matrix = np.ones((num_categories, num_categories))

    # Perform one-way ANOVA for all pairs of samples
    for i, j in combinations(range(num_categories), 2):
        category1, category2 = categories[i], categories[j]
        sample1 = data.loc[data[category_col] == category1][value_column]  # Replace 'value_column' with the actual numerical variable column
        sample2 = data.loc[data[category_col] == category2][value_column]

        _, p_value = f_oneway(sample1, sample2)

        # Store p-value in the matrix
        p_values_matrix[i, j] = p_value
        p_values_matrix[j, i] = p_value  # Since the matrix is symmetric

    # Create a DataFrame for the heatmap
    columns = [f"{category}" for category in categories]
    p_values_df = pd.DataFrame(p_values_matrix, index=columns, columns=columns)

    # Plot the heatmap
    plt.figure(figsize=(15, 12))  # Adjust the figure size as needed
    sns.heatmap(p_values_df, annot=True, cmap=plt.cm.get_cmap('coolwarm').reversed(), vmin=0, vmax=alpha, fmt=".3f",**args)
    plt.title("One-way ANOVA P-Values Heatmap")
    plt.show()

# Example usage:
# Assuming you have a DataFrame named 'df' with columns 'category_column' and 'value_column'
# Replace 'category_column' and 'value_column' with the actual column names in your dataset
```

```python
one_way_anova_heatmap_from_dataframe(data=data[["price","condition"]], category_col='condition',value_column="price")
```
